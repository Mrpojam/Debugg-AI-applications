# Read data

```python
import pandas as pd
import numpy as np
```

```python
train_data = pd.read_csv("/content/Corona_NLP_train.csv",encoding='latin-1')
train_data.head()
```

```python
y = pd.get_dummies(train_data["Sentiment"])
y
```

```python
train_data.isnull().sum()
```

```python
def data_pre (data):
  data.drop("UserName", axis=1, inplace=True)
  data.drop("ScreenName", axis=1, inplace=True)
  data.drop("Location", axis=1, inplace=True)
  data.drop("TweetAt", axis=1, inplace=True)
  y = pd.get_dummies(train_data["Sentiment"])
  data.drop("Sentiment", axis=1, inplace=True)
  data = pd.concat([data, y], axis=1)
  return data
  
```

```python
train_data = data_pre(train_data)
train_data.head()
```
